```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AutoMPG.csv')
```

```python
df
```

```python
df = df.drop(['car name'],axis=1)
```

```python
X = df.drop(['mpg'],axis=1)
Y = df['mpg']
```

```python
model = sm.OLS(Y,sm.add_constant(X['weight'])).fit()
```

```python
model.summary()
```

```python
plt.plot(df['weight'],df['mpg'],'ro')
plt.plot(df['weight'],model.fittedvalues,  'b')
plt.legend(['Data', 'Fitted model'])
plt.xlabel('weight')
plt.ylabel('mpg')
```

```python
modelall = sm.OLS(Y,sm.add_constant(X)).fit()
```

```python
modelall.summary()
```

## Another Function for Linear Regression

```python
model2 = ols("mpg ~ weight", data=df).fit()
```

```python
model2.summary()
```

## Seaborn to plot Regression Line

```python
#Seaborn
sns.lmplot(x='weight', y='mpg', data=df)
```

### One Hot Encoding - for categorical columns

```python
df = pd.read_csv('AutoMPG.csv')
```

```python
df['Type'] = 'Old'
```

```python
df.loc[df['model year'] > 70, 'Type'] = 'New'
```

```python
df = df.drop('car name',axis=1)
```

```python
df
```

```python
X = df.drop(['mpg'],axis=1)
Y = df['mpg']
```

```python
modelcat = sm.OLS(Y,sm.add_constant(X)).fit()
```

```python
one_hot = pd.get_dummies(df, columns = ['Type'])
```

```python
one_hot
```

```python
X = one_hot.drop(['mpg'],axis=1)
Y = one_hot['mpg']
```

```python
modelcat = sm.OLS(Y,sm.add_constant(X)).fit()
```

```python
modelcat.summary()
```
