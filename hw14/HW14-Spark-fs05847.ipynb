## 1. Find the names and weight of the top 15 products by weight. Sort weight by descending order.

### Using RDD

```python
(content.filter(lambda line: line.split("\t")[7] != "NULL")
 .map(lambda line: (line.split("\t")[1], float(line.split("\t")[7])))
                     .takeOrdered(15, lambda x : -x[1])
                    )
```

### Using Spark DataFrame

```python
# selecting two columns Name and Weight from original dataframe and convert into their proper datatype. Default is string
df1 = df.select(df.Name, df.Weight.cast("float"))
# df1.sort(df.desc)
df1.orderBy('Weight', ascending=False).show(15, truncate=False)
```

### Using Spark SQL

```python
# convert the above Dataframe into SQL Table
df1.createOrReplaceTempView("df_table")
sqlContext.sql(" SELECT Name, Weight FROM df_table ORDER BY Weight DESC limit 15").show(truncate=False)
```

## 4. Retrieve the product name, product number, color and list price of products whose product number begins with 'BK-'. Retrieve only 15 records and Sort the price in descending order

### Using RDD

```python
(content.filter(lambda line: "BK" in line.split("\t")[2])
 .map(lambda line: (line.split("\t")[1], line.split("\t")[2], line.split("\t")[3], float(line.split("\t")[5])))
.takeOrdered(15, lambda x: -x[3]))
```

### Using Spark Dataframe

## 5. Visualize Average List Price per Color and Average List Price per Size. Discuss/Explain your results.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from pyspark.sql import functions as F
%matplotlib inline
```

```python
df1 = df.select(df.Color, df.ListPrice.cast("float"))
lst = df1.groupBy("Color").avg("ListPrice")
colors = lst.rdd.map(lambda x: x.Color).collect()
prices = lst.rdd.map(lambda x: x['avg(ListPrice)']).collect()
plt.ylabel("Avg ListPrice")
plt.bar(colors, prices)
plt.xlabel("Color")
plt.show()
```

### x-axis = Color, Y-axis= avg ListPrice
As we can see from the above the graph that Red has the highest Average ListPrice Red. Blue has has second highest and Yellow has third highest. And Yellow and blue are very close to each other.
White,Multi color has very low Average listPrice, less than 50.

```python
df1 = df.select(df.Size, df.ListPrice.cast("float"))
lst = df1.groupBy("Size").avg("ListPrice")
size = lst.rdd.map(lambda x: x.Size).collect()
prices = lst.rdd.map(lambda x: x['avg(ListPrice)']).collect()
plt.ylabel("Avg ListPrice")
plt.bar(size, prices)
plt.xlabel("Size")
plt.show()
```

### x-axis = Size, Y-axis= avg ListPrice
As we can see from the above the graph that Size 56 has the highest Average ListPrice. Size 38 has has second highest Average ListPrice and 46 has third highest Average ListPrice. 46 and 38 are very close to each other.
Size M, L, XL has very low Average listPrice.

## 5. Visualize Total Count per Size. Discuss/Explain your results.

```python
df1 = df.select(df.Name, df.ListPrice, df.Color, df.Size)
df1.createOrReplaceTempView("df_t")

re = df1.groupby("Size").agg(F.count("ListPrice")).collect()
price = []
size = []

for i in re:
  price.append(i[1])
  size.append(i[0])
 
plt.bar(size, price)
plt.ylabel("count")
plt.xlabel("Size")
plt.show()
```

As we can see from the above the graph that Size "NULL' has the highest count in our dataset. Size 44 has has second highest count and 48 has third highest count. if we ignore the NULL size then size 44 has largest frequency around 30 in our dataset. 
Size 56, 70 has very low count close to 0

# Using Machine Learning Library (For Practice)

```python
dfall = df.select(df.Weight.cast("float"), df.Size.cast("float"), df.ListPrice.cast("float"))
dfall.show(5, truncate=False)
```

```python
dfall = dfall.na.drop("any")
```

```python
feature_columns = dfall.columns[:-1] # here we omit the final column for prediction
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=feature_columns,outputCol="features")
```

```python
data_2 = assembler.transform(dfall)
```

```python
data_2.show()
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
algo = LinearRegression(featuresCol="features", labelCol="ListPrice")
```

```python
model = algo.fit(data_2)
```

```python
evaluation_summary = model.evaluate(data_2)
```

```python
evaluation_summary.rootMeanSquaredError
```
